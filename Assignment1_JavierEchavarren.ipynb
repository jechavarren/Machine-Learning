1.  Explain the differences between supervised, semi-surpervised, weakly-supervised and unsupervised learning.




**Answer:** 

Supervised learning is a type of machine learning algorithm where a model is trained on labeled data that has been pre-classified. The objective of the model is to predict the class of the new data points based on their attributes, and following the knowledge obtained by the training data.

Semi-supervised learning is a type of supervised learning where only part of the data is labeled. The unlabeled data is used to extract features about the data and the labeled data to train the prediction capabilities. The final objective is predicting the class or other attribute of new data, as with supervised learning.

Weakly-supervised learning is a type of supervised learning where the data labels are either in low quantity or low quality. These models try to predict the class of new datapoints, but many times these models cannot generalize well.

Unsupervised learning is a type of machine learning that uses unlabeled data to identify patterns and features in the data that can be used to group the data into different categories.

3. Explain how to perform N-fold cross-validation.




**Answer:**

N-fold cross validation is a technique used to evaluate the performance of a machine learning model, maximizing the amount of data used for training but allowing the training to still be evaluated independently.

To perform N-fold cross-validation, we divide the dataset into N parts. We then use N-1 parts for training and the other part for testing. We perform this method N times choosing each time a different part for the testing, and we finally evaluate the model performance by averaging the performance of the N iterations.

```python
# define cost function 
def mse(predictions, results):
  difference = predictions-results
  return sum(difference**2)/len(difference)


```

**Question: How did you choose your lr and epoch number?**

**Answer:**

```python
# gradient descent to find the optimal fit
# Initialize learning rate to 0.5, as I want it not to be very slowly but also not to oscillate too much.
# Initialize the epoch vector as range from 1 to 1000, as this program does not consume much and I want to have margin for the model to converge in case it is slow to converge.
learning_rate = 0.5
epoch = range(1,1000)

batch_size = 100

#Construct Gradient descent function
def gradient_descent(theta, learning_rate, X,y, history_cost):
  f = y - (theta[0] + theta[1]*X )
  N = len(X)
  theta[0] -= learning_rate* (-2*f.sum()/N)
  theta[1] -= learning_rate * (-2*sum(X*f) /N)
  predictions = theta[0] + theta[1]*X
  error = mse (predictions, y)
  history_cost.append(error)


  return history_cost, theta

history_cost=[]

for i in epoch:
  indexes = np.random.randint(0, len(X), batch_size)
  X1 = np.take(X, indexes)
  y1 = np.take(y, indexes)
  history_cost, theta =gradient_descent(theta, learning_rate, X1, y1, history_cost)

```

```python
# draw the cost change 
plt.plot(epoch, history_cost)

plt.xlabel("Epochs")
plt.ylabel("Mean Square Error")
plt.title("Evolution of MSE between epochs")

plt.show()
```
